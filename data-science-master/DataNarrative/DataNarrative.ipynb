```python
import pandas as pd

df = pd.read_csv('source/car_sales.csv')
df.head()
```

```python
df.set_index('DATE',inplace=True)
```

```python
import matplotlib.pyplot as plt

df.plot()
plt.grid()
plt.title('Quarterly Car Sales by Dealership', color='black')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout()
plt.savefig('plots/raw.png')
plt.show()
```

```python
others = [c for c in df.columns if c != 'LAKESIDE']
others
```

```python
plt.plot(df['LAKESIDE'], color='red', label='LAKESIDE')
for c in others:
    plt.plot(df[c], color='grey', label=c)
plt.grid()
plt.xticks(rotation=45)
plt.title('Quarterly Car Sales by Dealership', color='black')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout()
plt.savefig('plots/lakeside.png')
plt.show()
```

```python
df.loc['2018 Q1']
```

```python
df.loc['2018 Q1'].plot(kind='bar')
plt.grid()
plt.xticks(rotation=45)
plt.title('Quarterly Car Sales by Dealership', color='black')
plt.tight_layout()
plt.savefig('plots/2018-q1.png')
plt.show()
```

```python
df.loc['2018 Q2'].plot(kind='bar')
plt.grid()
plt.title('Quarterly Car Sales by Dealership', color='black')
plt.tight_layout()
plt.savefig('plots/2018-q2.png')
plt.show()
```

```python
colormap = []
for c in df.columns:
    if c == 'LAKESIDE':
        colormap.append('red')
    else:
        colormap.append('grey')
(df.loc['2018 Q2']-df.loc['2018 Q1']).plot(kind='bar',color=colormap)
plt.grid()
plt.title('Difference between the second and first quarters of 2018', color='black')
plt.tight_layout()
plt.savefig('plots/diff.png')
plt.show()
```
